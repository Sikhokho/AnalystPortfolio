```python
#rename columns
newHeadingsDict = {
    "timestamp" : "Time",
    "cnt" : "Count",
    "t1" : "RealTemp",
    "t2" : "FeelsLikeTemp",
    "hum" : "Humidity",
    "wind_speed" : "WindSpeed (km/h)",
    "weather_code" : "Weather",
    "is_holiday" : "isHoliday",
    "is_weekend" : "isWeekend",
    "season" : "Season"
}

bikes.rename(newHeadingsDict, axis = 1, inplace = True)
bikes
```

```python
#convert humidity column to percentages
bikes.Humidity = bikes.Humidity / 100
bikes
```

```python
#Applying the correct values according to the codes provided in the dataset's documentation
seasonDict = {
    "3.0":"Winter",
    "2.0":"Fall",
    "1.0":"Summer",
    "0.0":"Spring",
}

weatherDict = {
    "1.0":"Clear",
    "2.0":"Scattered Clouds",
    "3.0":"Broken Clouds",
    "4.0":"Cloudy",
    "7.0":"Rain",
    "10.0":"Rain & Thunderstorm",
    "26.0":"Snowfall",
    "94.0":"Freezing Fog",
}

bikes.Season = bikes.Season.astype('str')
bikes.Season = bikes.Season.map(seasonDict)

bikes.Weather = bikes.Weather.astype('str')
bikes.Weather = bikes.Weather.map(weatherDict)
bikes.info()
```
