```python
df.sort_values(by = "World Population Percentage", ascending = False).head(10)
# sorting the dataframe by 'World Population Percentage' column in descending order and then taking the top 10 rows
```

```python
df.corr(numeric_only = True) #shows the correlation between each pair of columns in the dataframe
```

```python
sns.heatmap(df.corr(numeric_only = True), annot = True)
plt.rcParams['figure.figsize'] = (10,7)
plt.show()
# visualisation of the correlation matrix with a colorbar showing the strengths of correlations
```

```python
df.groupby(by = "Continent").mean(numeric_only = True).sort_values(by="2022 Population", ascending = False)
#grouping  by Continent and calculating the mean of each column for 2022 Population only, then sorting in descending order based on this new population column
```

```python
df[df['Continent'].str.contains('Oceania')]
#finding the data for countries in Oceania
```

```python
df2 = df.groupby(by = "Continent")[df.columns[5:13]].mean(numeric_only = True).sort_values(by="2022 Population", ascending = False)
df2
#creating a 2nd  dataframe with the same columns as df1 only containing the mean population per year for every continent
```

```python
df3 = df2.transpose()
df3
#creating a 3rd dataframe with the same data as df2 with column headings swapped
```

```python
df3.plot(figsize = (20,10))
#line chart showing the data from df3
```
